```python
import math
from math import cos, sin
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rotZ(t):
    return (np.array([[cos(t), -sin(t)], [sin(t), cos(t)]]))
def cross(x, y):
    return(np.dot(x, y))
```

```python
p1 = np.array([1.0, 5.0])
p2 = np.array([4.0, 1.0])
a = 1
b = 2

x = np.arange(0, 5, 0.1)
r = p2 - p1
y = (r[1]/r[0]) *(x - p2[0]) + p2[1]

theta = math.atan(r[1]/r[0])

r2 = cross(r, rotZ(theta))
p1r = cross(p1, rotZ(theta))
p2r = cross(p2, rotZ(theta))
r2 = p2r - p1r

y2 = (r2[1]/r2[0]) *(x - p2[0]) + p2[1]

plt.plot(y, x)
plt.legend('Origen')
plt.show()

plt.plot(y2, x)
plt.show()
```
